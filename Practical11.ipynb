```python
import nltk

# Download required resources (run once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')  # IMPORTANT
```

```python
from nltk.tokenize import word_tokenize
from nltk import pos_tag
from nltk.chunk import RegexpParser
```

```python
text = """
Natural language processing is a powerful technique.
It helps machines understand human language and extract useful information.
"""
```

```python
tokens = word_tokenize(text)
print("Tokens:\n", tokens)
```

```python
pos_tags = pos_tag(tokens)
print("\nPOS Tags:\n", pos_tags)
```

```python
grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
```

```python
chunk_parser = RegexpParser(grammar)
tree = chunk_parser.parse(pos_tags)

print("\nChunk Tree:\n", tree)
```

```python
print("\nNoun Phrases:")

for subtree in tree.subtrees():
    if subtree.label() == "NP":
        np = " ".join(word for word, tag in subtree)
        print(np)
```

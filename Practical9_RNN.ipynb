```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
text_data = """
I love machine learning and artificial intelligence
machine learning is used in many applications
deep learning is a subset of machine learning
artificial intelligence is the future of technology
I enjoy learning new things in data science
data science involves statistics and programming
python is widely used for machine learning
neural networks are powerful models for learning
"""
```

```python
def preprocess_text(text):
    text = text.lower()
    return text
```

```python
def tokenize_text(text):
    tokenizer = Tokenizer()
    tokenizer.fit_on_texts([text])

    total_words = len(tokenizer.word_index) + 1

    return tokenizer, total_words
```

```python
def create_sequences(tokenizer, text, seq_len):
    input_sequences = []

    token_list = tokenizer.texts_to_sequences([text])[0]

    for i in range(seq_len, len(token_list)):
        n_gram_seq = token_list[i-seq_len:i+1]
        input_sequences.append(n_gram_seq)

    input_sequences = np.array(input_sequences)

    X = input_sequences[:, :-1]
    y = input_sequences[:, -1]

    return X, y
```

```python
def build_model(total_words, seq_len):
    model = Sequential()

    model.add(Embedding(total_words, 64, input_length=seq_len))
    model.add(LSTM(100))
    model.add(Dense(total_words, activation='softmax'))

    model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

```python
def train_model(model, X, y):
    model.fit(X, y, epochs=80, verbose=1)
```

```python
def generate_text(model, tokenizer, seed_text, seq_len, num_words):
    for _ in range(num_words):
        token_list = tokenizer.texts_to_sequences([seed_text])[0]
        token_list = pad_sequences([token_list], maxlen=seq_len, padding='pre')

        predicted = np.argmax(model.predict(token_list), axis=-1)

        for word, index in tokenizer.word_index.items():
            if index == predicted:
                seed_text += " " + word
                break

    return seed_text
```

```python
# Preprocess
text = preprocess_text(text_data)

# Tokenize
tokenizer, total_words = tokenize_text(text)

# Create sequences
seq_len = 3
X, y = create_sequences(tokenizer, text, seq_len)

# Build model
model = build_model(total_words, seq_len)

# Train
train_model(model, X, y)

# Generate text
while True:
    seed = input("Enter seed text (or 'exit'): ")

    if seed.lower() == "exit":
        break

    output = generate_text(model, tokenizer, seed, seq_len, 5)

    print("Generated Text:", output)
```

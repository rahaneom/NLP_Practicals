**Paragraph -> Corpus**
**Sentence -> Document**
**Unique Words - > Vocabulary**

1. **Tokenization -> punkt_tab**
  From Sentence to Words -> word_tokenize()
  From Corpus to Sentence -> sent_tokenize()

```python
import nltk
nltk.download("punkt_tab")
```

```python
sentence = """The Supreme Court on Monday didn't the reinstatement of a Madhya Pradesh judge who was dismissed from service in 2019 for allegedly misbehaving with a woman co passenger in a train while in an inebriated state, with the top court describing the case as “shocking”, “disgusting” and observing that “nothing less than termination” was warranted in such circumstances."""
```

```python
tokens = nltk.word_tokenize(sentence)
```

```python
tokens
```

```python
sentence
```

```python
nltk.sent_tokenize(sentence)
```

**2. Stemming -> Finding Root Word using PorterStemmer**

**Method -> ps.stem(each word in tokens array)**

Types:
Porter Stemmer, Snowball Stemmer, Regexp Stemmer, Lancaster Stemmer

```python
from nltk.stem import PorterStemmer
ps = PorterStemmer()
```

```python
for t in tokens:
  print(t, " : ", ps.stem(t))
```

**3. Lemmatization: Download wordnet from nltk and use WordNetLemmatizer**

**Method: lemma.lemmatize(each word in tokens array)**

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemma = WordNetLemmatizer()
```

```python
for t in tokens:
  print(t, " : ", lemma.lemmatize(t))
```

4. Identify & Remove Stopwords
Download stopwords from nltk
Method -> stopwords.words(Language)

```python
from nltk.corpus import stopwords
nltk.download("stopwords")

sw = set(stopwords.words("english"))

filtered_tokens = [t for t in tokens if t.lower() not in sw]

print(filtered_tokens)
```

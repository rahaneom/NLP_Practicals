```python
import nltk
nltk.download('punkt_tab')
```

```python
from nltk.tokenize import word_tokenize
from nltk import ngrams
```

```python
sentence = "I am performing Natural Language Processing Practical No. 3"
```

```python
sentence = sentence.lower()
tokens = word_tokenize(sentence)
```

```python
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
lemma_tokens = [lemmatizer.lemmatize(token) for token in tokens if token not in stop_words]
lemma_tokens
```

```python
bigrams = list(ngrams(lemma_tokens, 2))
print(bigrams)
```

```python
trigrams = list(ngrams(lemma_tokens, 3))
print(trigrams)
```

```python
quadgrams = list(ngrams(lemma_tokens, 4))
print(quadgrams)
```

```python
def get_ngrams(n, tokens):
    ngrams = []
    tokens = (n-1)*['<START>']+tokens
    for i in range(n - 1, len(tokens)):
        context = tuple(tokens[i - p - 1] for p in reversed(range(n - 1)))
        target_word = tokens[i]
        ngrams.append((context, target_word))
    return ngrams
```

```python
trigrams = get_ngrams(3, lemma_tokens)
print(trigrams)
```

```python
for context, word in trigrams:
    print(context, word)
```

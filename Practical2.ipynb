```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag

nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
text = "NLTK is a powerful library for natural language processing. Parts of Speech (PoS) tagging is a fundamental task in Natural Language Processing (NLP) where each word in a sentence is assigned a grammatical category such as noun, verb, adjective or adverb. This process help machines to understand the structure and meaning of sentences by identifying the roles of words and their relationships."
words = word_tokenize(text)

pos_tags = pos_tag(words)
```

```python
print("Original Text:")
print(text)
```

```python
print("\nPoS Tagging Result:")

for word, tag in pos_tags:
    print(f"{word} : {tag}")
```
